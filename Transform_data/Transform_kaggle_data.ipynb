# Clean the Kaggle Data

## Initial Look at the Movie Metadata

```python
# Check is that all of the columns came in as the correct data types
kaggle_metadata.dtypes
```

```python
# Check that all the values are either True or False for "adult" and "video"
kaggle_metadata['adult'].value_counts()
```

## Remove Bad Data

```python
# Remove Bad Data
kaggle_metadata[~kaggle_metadata['adult'].isin(['True','False'])]
```

```python
# Drop bad data and adult movies
kaggle_metadata = kaggle_metadata[kaggle_metadata['adult'] == 'False'].drop('adult',axis='columns')
```

```python
# Look at the values of the video column
kaggle_metadata['video'].value_counts()
```

## Convert Data Types

```python
# Convert Boolean column
kaggle_metadata['video'] == 'True'
```

```python
# Assign Boolean column back to video
kaggle_metadata['video'] = kaggle_metadata['video'] == 'True'
```

```python
# Check if there's any data that can't be converted to numbers
# Use the to_numeric() method from Pandas.
# Make sure the 'errors='' argument is set to 'raise'
kaggle_metadata['budget'] = kaggle_metadata['budget'].astype(int)
kaggle_metadata['id'] = pd.to_numeric(kaggle_metadata['id'], errors='raise')
kaggle_metadata['popularity'] = pd.to_numeric(kaggle_metadata['popularity'], errors='raise')
```

```python
# Convert release_date to datetime, by using Pandas's built-in function 'to_datetime()'
kaggle_metadata['release_date'] = pd.to_datetime(kaggle_metadata['release_date'])
```

```python
# Save Kaggle_metadata as CSV file
df = pd.DataFrame(kaggle_metadata)
df.to_csv('../Resources_clean/ETL_kaggle_metadata.csv')
```
